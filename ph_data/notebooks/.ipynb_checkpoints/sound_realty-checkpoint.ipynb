```python
# Modal app
app = App("sr-hybrid-upload-app")

# Persistent Modal volume
volume = Volume.from_name("sr-hybrid-app-volume", create_if_missing=True)

# Image with boto3 for S3 + optional baked-in dataset + baked-in model artifacts
image = (
    Image.debian_slim()
    .pip_install("boto3")
    .add_local_dir("../data", "/frozen_data")               # dataset (optional frozen)
    .add_local_dir("../model", "/frozen_model")        # model artifacts (optional frozen)
)


@app.function(volumes={"/data": volume}, image=image)
def upload_all(local_dirs: dict = None, s3_bucket: str = None, s3_prefix: str = None, use_frozen: bool = False):
    """
    Uploads training data + model artifacts into the Modal volume.

    - local_dirs: dict of {"remote_subdir": "local_path"} (dev mode)
    - s3_bucket + s3_prefix: fetch from S3 (prod mode)
    - use_frozen: copy pre-baked datasets + models
    """
    os.makedirs("/data", exist_ok=True)

    if local_dirs:
        # Dev mode
        for subdir, local_path in local_dirs.items():
            dest_dir = Path(f"/data/{subdir}")
            dest_dir.mkdir(parents=True, exist_ok=True)

            for file in Path(local_path).glob("*"):
                if file.is_file():
                    shutil.copy(file, dest_dir / file.name)
                    print(f"[DEV] Copied {file} -> {dest_dir / file.name}")

    elif s3_bucket and s3_prefix:
        # Prod mode (fetching from S3)
        s3 = boto3.client("s3")
        result = s3.list_objects_v2(Bucket=s3_bucket, Prefix=s3_prefix)

        for obj in result.get("Contents", []):
            key = obj["Key"]
            filename = os.path.basename(key)
            subdir = os.path.dirname(key).split("/")[-1]  # e.g. "model" or "data"
            dest_dir = Path(f"/data/{subdir}")
            dest_dir.mkdir(parents=True, exist_ok=True)

            dest = dest_dir / filename
            s3.download_file(s3_bucket, key, str(dest))
            print(f"[PROD] Downloaded s3://{s3_bucket}/{key} -> {dest}")

    elif use_frozen:
        # Frozen mode (both datasets + model artifacts baked in)
        for folder, frozen_path in [("data", "/frozen_data"), ("model", "/frozen_model")]:
            dest_dir = Path(f"/data/{folder}")
            dest_dir.mkdir(parents=True, exist_ok=True)

            for file in Path(frozen_path).glob("*"):
                if file.is_file():
                    shutil.copy(file, dest_dir / file.name)
                    print(f"[FROZEN] Copied {file} -> {dest_dir / file.name}")

    else:
        print("⚠️ No source provided. Pass local_dirs, or s3_bucket+s3_prefix, or use_frozen=True.")

    # Confirm
    print("\nFiles now in Modal volume:")
    for file in Path("/data").rglob("*"):
        print(f" - {file}")


# === Notebook/CLI helpers ===

def run_upload_local():
    """Upload local dataset + model artifacts (dev mode)."""
    local_dirs = {
        "data": "../data",                  # training data
        "model": "../model"            # model artifacts (pkl, json, etc.)
    }
    with app.run():
        upload_all.remote(local_dirs=local_dirs)


def run_upload_s3(bucket, prefix):
    """Upload from S3 (prod mode)."""
    with app.run():
        upload_all.remote(s3_bucket=bucket, s3_prefix=prefix)


def run_upload_frozen():
    """Upload frozen dataset + model artifacts (frozen mode)."""
    with app.run():
        upload_all.remote(use_frozen=True)
```

## Model Serving

```python
# Cell 1: Imports and setup
import modal
import pandas as pd
import numpy as np
from fastapi import File, UploadFile, Form, HTTPException
import io
import pickle
import os

# Cell 2: App definition and image setup
app = modal.App("sr-hybrid-model-api")

base_image = (modal.Image.debian_slim()
        .pip_install(["fastapi","uvicorn","scikit-learn==1.7.2","pandas","numpy","joblib","bentoml","dash","matplotlib","seaborn","tqdm","wandb"]))

fastai_image = (base_image.pip_install(["fastai", "torch"]))
model_volume = modal.Volume.from_name("sr-hybrid-app-volume")

# Cell 3: Define your functions
def load_model(model_path="/vol/model/model.pkl"):
    """Function to load the model from Modal volume"""
    if os.path.exists(model_path):
        print(f"Loading existing model from pickle at {model_path}")
        with open(model_path, 'rb') as f:
            model = pickle.load(f)
        return model
    else:
        print(f"Model file not found at {model_path}")
        return None

@app.function(image=fastai_image, volumes={"/vol": model_volume})
def serve_model():
    """Load your pre-trained model from the sr-hybrid-app-volume"""
    try:
        print("Loading model from sr-hybrid-app-volume...")
        model = load_model()
        if model is None:
            raise Exception("Failed to load model from volume")
        print("Model loaded successfully!")
        return model
    except Exception as e:
        import traceback
        print(f"Error loading model: {str(e)}")
        print(traceback.format_exc())
        raise

@app.function(image=base_image)
@modal.fastapi_endpoint(method="GET")
def health():
    """Health check endpoint"""
    return {"status": "healthy", "service": "sr-hybrid-model-api"}

@app.function(image=fastai_image, volumes={"/vol": model_volume})
@modal.fastapi_endpoint(method="POST")
async def predict_csv(file: UploadFile = File(...)):
    """API endpoint for batch predictions from a CSV file"""
    try:
        model = serve_model.remote()
        contents = await file.read()
        
        try:
            test_df = pd.read_csv(io.BytesIO(contents))
        except Exception as e:
            return {"success": False, "error": f"Failed to parse uploaded CSV: {str(e)}"}
        
        # Add your preprocessing here
        predictions = model.predict(test_df)
        
        return {
            "success": True,
            "predictions": predictions.tolist() if hasattr(predictions, 'tolist') else predictions
        }
        
    except Exception as e:
        import traceback
        return {
            "success": False,
            "error": f"Error processing CSV: {str(e)}",
            "traceback": traceback.format_exc()
        }


# Cell 4: Test model loading with app context
print("Testing model loading...")

# The key is to run within the app context
with app.run():
    result = serve_model.remote()
    print("Model loaded successfully!")
    print(f"Model type: {type(result)}")
```

```python
# Cell: Deploy enhanced API with demographic data preprocessing
import subprocess
import tempfile
import os

# Complete app code with demographic preprocessing
app_code = '''
import modal
import pandas as pd
import numpy as np
from fastapi import File, UploadFile, Form, HTTPException
import io
import pickle
import os
import json
from pathlib import Path

app = modal.App("sr-hybrid-model-api")
base_image = (modal.Image.debian_slim()
        .pip_install(["fastapi","uvicorn","scikit-learn==1.7.2","pandas","numpy","joblib","bentoml","dash","matplotlib","seaborn","tqdm","wandb"]))
fastai_image = (base_image.pip_install(["fastai", "torch"]))
model_volume = modal.Volume.from_name("sr-hybrid-app-volume")

def load_model(model_path="/vol/model/model.pkl"):
    """Function to load the model from Modal volume"""
    if os.path.exists(model_path):
        print(f"Loading existing model from pickle at {model_path}")
        with open(model_path, 'rb') as f:
            model = pickle.load(f)
        return model
    else:
        print(f"Model file not found at {model_path}")
        return None

@app.function(image=fastai_image, volumes={"/vol": model_volume})
def serve_model():
    """Load your pre-trained model from the sr-hybrid-app-volume"""
    try:
        print("Loading model from sr-hybrid-app-volume...")
        model = load_model()
        if model is None:
            raise Exception("Failed to load model from volume")
        print("Model loaded successfully!")
        return model
    except Exception as e:
        import traceback
        print(f"Error loading model: {str(e)}")
        print(traceback.format_exc())
        raise

@app.function(image=fastai_image, volumes={"/vol": model_volume})
def load_model_features():
    """Load expected model features from volume"""
    features_path = Path("/vol/model/model_features.json")
    if features_path.exists():
        print(f"Loading model features from: {features_path}")
        return json.loads(features_path.read_text())
    else:
        # Fallback to hardcoded features if file not found
        print("Model features file not found, using fallback")
        return ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement', 'ppltn_qty', 'urbn_ppltn_qty', 'sbrbn_ppltn_qty', 'farm_ppltn_qty', 'non_farm_qty', 'medn_hshld_incm_amt', 'medn_incm_per_prsn_amt', 'hous_val_amt', 'edctn_less_than_9_qty', 'edctn_9_12_qty', 'edctn_high_schl_qty', 'edctn_some_clg_qty', 'edctn_assoc_dgre_qty', 'edctn_bchlr_dgre_qty', 'edctn_prfsnl_qty', 'per_urbn', 'per_sbrbn', 'per_farm', 'per_non_farm', 'per_less_than_9', 'per_9_to_12', 'per_hsd', 'per_some_clg', 'per_assoc', 'per_bchlr', 'per_prfsnl']

@app.function(image=fastai_image, volumes={"/vol": model_volume})  
def load_demographics_data():
    """Load demographics data from volume"""
    # Check possible locations for demographics data
    possible_paths = [
        Path("/vol/data/zipcode_demographics.csv"),
        Path("/vol/zipcode_demographics.csv"),
        Path("/vol/demographics.csv")
    ]
    
    for path in possible_paths:
        if path.exists():
            print(f"Loading demographics data from: {path}")
            demographics_df = pd.read_csv(path)
            # Ensure zipcode is string to preserve leading zeros
            demographics_df['zipcode'] = demographics_df['zipcode'].astype(str)
            return demographics_df
    
    raise FileNotFoundError("Demographics data not found in volume. Expected at /vol/data/zipcode_demographics.csv")

@app.function(image=fastai_image, volumes={"/vol": model_volume})
def preprocess_for_inference(test_df, demographics_df, model_expected_features):
    """Preprocess input data by merging with demographics and preparing for model"""
    
    # Ensure zipcode is string in test data
    test_df['zipcode'] = test_df['zipcode'].astype(str)
    
    # Merge with demographics data on zipcode
    print("Merging test data with demographics...")
    inference_df = test_df.merge(demographics_df, on='zipcode', how='left')
    
    print(f"Original test data shape: {test_df.shape}")
    print(f"After demographics merge shape: {inference_df.shape}")
    
    # Check which expected features are missing
    missing_features = set(model_expected_features) - set(inference_df.columns)
    if missing_features:
        print(f"WARNING: Missing expected features: {missing_features}")
        
        # Add missing features with default values
        for feature in missing_features:
            inference_df[feature] = 0.0
            print(f"Added missing feature '{feature}' with default value 0.0")
    else:
        print("All expected features are present after merge.")
    
    # Handle missing values with median imputation
    for col in model_expected_features:
        if col in inference_df.columns and inference_df[col].isnull().any():
            median_val = inference_df[col].median()
            if pd.isna(median_val):  # If all values are NaN, use 0
                median_val = 0.0
            inference_df[col] = inference_df[col].fillna(median_val)
            print(f"Imputed NaNs in '{col}' with median: {median_val}")
    
    # Select only the expected features in the correct order
    final_df = inference_df[model_expected_features]
    
    print(f"Final processed data shape: {final_df.shape}")
    print(f"Final features: {list(final_df.columns)}")
    
    return final_df

@app.function(image=base_image)
@modal.fastapi_endpoint(method="GET")
def health():
    """Health check endpoint"""
    return {"status": "healthy", "service": "sr-hybrid-model-api"}

@app.function(image=fastai_image, volumes={"/vol": model_volume})
@modal.fastapi_endpoint(method="POST")
async def predict_csv(file: UploadFile = File(...)):
    """API endpoint for batch predictions with automatic demographic data merging"""
    try:
        # Load model
        model = serve_model.remote()
        
        # Load model features configuration
        model_features = load_model_features.remote()
        
        # Load demographics data 
        demographics_df = load_demographics_data.remote()
        
        # Parse uploaded CSV
        contents = await file.read()
        try:
            test_df = pd.read_csv(io.BytesIO(contents))
            print(f"Input CSV columns: {list(test_df.columns)}")
            print(f"Input CSV shape: {test_df.shape}")
        except Exception as e:
            return {"success": False, "error": f"Failed to parse uploaded CSV: {str(e)}"}
        
        # Preprocess the data (merge with demographics and prepare for model)
        processed_df = preprocess_for_inference.remote(test_df, demographics_df, model_features)
        
        # Make predictions
        predictions = model.predict(processed_df)
        
        return {
            "success": True,
            "predictions": predictions.tolist() if hasattr(predictions, 'tolist') else predictions,
            "processed_features": processed_df.columns.tolist(),
            "input_shape": test_df.shape,
            "processed_shape": processed_df.shape,
            "num_predictions": len(predictions)
        }
        
    except Exception as e:
        import traceback
        return {
            "success": False,
            "error": f"Error processing CSV: {str(e)}",
            "traceback": traceback.format_exc()
        }
'''

# Write to temporary file and deploy
with tempfile.NamedTemporaryFile(mode='w', suffix='.py', delete=False) as f:
    f.write(app_code)
    temp_file = f.name

try:
    # Deploy the app
    result = subprocess.run(['modal', 'deploy', temp_file], 
                          capture_output=True, text=True, check=True)
    print("Deployment successful!")
    print(result.stdout)
    
    print("\nYour API endpoints are now live and persistent!")
    print("Key features:")
    print("- Automatically loads demographics data from volume")
    print("- Merges input CSV with demographics based on zipcode")
    print("- Handles missing features and values")
    print("- Returns predictions for properly preprocessed data")
    
except subprocess.CalledProcessError as e:
    print("Deployment failed:")
    print(e.stderr)
finally:
    # Clean up temp file
    os.unlink(temp_file)
```
